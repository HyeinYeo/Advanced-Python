#### **Frequency Distribution**
- `단어의 빈도수`를 활용할 수 있는 많은 메소드 제공
- 단어 빈도수 중요성
    - 인간 언어 -> 컴퓨터가 인식
      - 전체 단어 수
      - 특정 단어의 빈도수

- **FreqDist(text1) : 딕셔너리 반환**

```python
FreqDist(text1) #이런 클래스로 되어있음
fdist = FreqDist(text1)
fdist #딕셔너리 = 키(단어):값(각 단어의 빈도수)
```

- **.most_common(n) : 최다 빈도수 단어 n개 출력**

```python
fdist.most_common(10) #가장 많이 쓴 단어 10개 출력 => 리스트로 출력
```

- **fdist[word] : 특정 단어의 빈도수 알고 싶을 때**

```python
fdist['whale'], text1.count('whale') #count()와 같음
```

- **.freq(word) : 특정 단어의 비율**

```python
fdist.freq('whale')
```

- **.plot(n, cumulative=True) : 최다 빈도수 순으로 50개 축적하여 시각화**

```python
fdist.plot(50, cumulative=True) #기울기가 완만해질수록 빈도 낮음
```

```python
fdist.plot(50, cumulative=False)
```

- **.hapaxes() : 최소 빈도수 = 한 번 나타나는 단어 반환**

```python
fdist.hapaxes() # list형으로 반환
```

```python
len(fdist.hapaxes()) #중복 제거된 상태, 대소문자는 구분되어 있는 상태(list형)
```

### **6.4 list comprehension**
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- `list + for_loop 한 줄로` 해결하는 방식
- `[A for A' in B]`
  - B에 있는 A'에서 A를 가진 `리스트`를 반환
  - 한 줄에 해결 가능한 방안
  - B: 기존 리스트
  - A': B의 자료 형태
  - A: 새로운 리스트의 요소(출력값)
- 조건+ : `[A for A' in B if <조건식>]`
  - [A for A' in B if <조건식> `and` <조건식>...]

```python
#for loop 돌면서 drinks에 있는 단어의 첫번째 원소만 뽑음
[d[0] for d in drinks] 
```

```python
# text1 알파벳 수가 10 이상인 단어
[word for word in text1 if len(word) >= 10]
```

```python
# text1 알파벳 수가 10 이상인 단어에서 맨 앞 글자만 반환
[word[0] for word in text1 if len(word) >= 10]
```

```python
# text4에서 'gnt' 포함하는 단어 반환
[word for word in text4 if word.find('gnt')!=-1]
[word for word in text4 if 'gnt' in word]        # 모범답안
```

```python
[word for word in set(text4) if 'gnt' in word] # 먼저 중복 없애고 추출
```

```python
set([word for word in text4 if 'gnt' in word]) # 추출하고 중복 없앰
```

```python
[word.lower() for word in set(text4) if 'gnt' in word] 
# 중복 제거 -> 조건에 맞는 단어 추출 -> 소문자로 변환
# 중복 남아있을 수 있음
```

### **출력 포매팅**
- 1. **`f'str... {var}'`**
  - f, F 둘 다 가능
- 2. **`%`** => 옛날 방식
  - '...  %자료형' %var
- 3. **`.format()`**
  - '... {}'.format(var)

```python
name = 'Son'
#print("my name is ###")

#1.
print(f'my name is {name}')

#2.
print("my name is %s" %name)

#3.
print("my name is {}".format(name))
```

```python
age = 14
name = 'leeeee'
#print("my name is ###")

#1.
print(f"my name is {name}, and I am {age} years old.")

#3.
print("my name is {:.2}, and I am {} years old.".format(name, age)) #문자열도 잘라짐
print("my name is {1}, and I am {0} years old.".format(name, age)) # 위치 지정도 가능
```

```python
# 소수점 아래 몇 자리 절삭
k = 0.555555
print("%.3f"%k)
```
